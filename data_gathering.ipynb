```python
import requests
import json
import pandas as pd
```

```python
url = "https://neuroelectro.org/api/1/nedm/"
neuro_data = requests.get(url).json()
```

```python
print(json.dumps(neuro_data, indent=4))
```

```python
# measurement type
neuro_data["objects"][0]["ecm"]["e"]["name"]
```

```python
#neuron type
neuro_data["objects"][0]["ncm"]
```

```python
#values
neuro_data["objects"][0]["val"]
```

```python
measurements = []
cells = []
values = []
types = []
for i in neuro_data["objects"]:
    measurements.append(i["ecm"]["e"]["name"])
    cells.append(i["ncm"]["n"]["name"])
    values.append(i["val"])
```

```python
measurements
cells
values
```

```python
data = list(zip(cells, values, measurements))
data
```

```python
df = pd.DataFrame(data,columns=["Cell Type", "Value", "Measurement"]) 
df
```
